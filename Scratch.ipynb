```python
import glob
import cv2
```

```python
GREYSCALE_ROOT = "/home/apil/data/hw_forms/test/images/"
```

```python
found_again = 0

for inFile in glob.glob(all_images_path+'*'):    
    imgID = inFile[inFile.rfind('/')+1 : inFile.rfind('.')]
    if imgID in new_images:
        im = cv2.imread(inFile, cv2.IMREAD_GRAYSCALE)        
        outfile = GREYSCALE_ROOT + imgID + ".jpg"
        cv2.imwrite(outfile, im)
```
